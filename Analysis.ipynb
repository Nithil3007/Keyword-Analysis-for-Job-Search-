```python
bigrams = 'bigrams_Output.txt'
file = open(bigrams, encoding="utf-8")
bigram = file.read()
input_string = bigram
bigram_list = ast.literal_eval(input_string)
```

```python
print(len(bigram_list))
bigrams = [tup for tup in bigram_list if tup[1] > 100]
print(len(bigrams))
```

```python
bigrams_trimmed = [tup for tup in bigram_list if tup[1] > 800]
print(len(bigrams_trimmed))
```

```python
bigrams_list = []
data_bigrams = []

for word in bigrams_trimmed:
    bigrams_list.append(word[0])

def bigram_freq_for_given_word(bigrams_list,bigrams,word):
    data_bigrams = []
    for i in range(len(bigrams_list)):
        if bigrams_list[i] ==  bigrams[i][0]:
            if bigrams[i][0].startswith(word):
                data_bigrams.append(bigrams[i])
    return data_bigrams

data_bigrams = bigram_freq_for_given_word(bigrams_list,bigrams,'data')
data_bigrams
```

```python
data = data_bigrams

# Separate the data into labels and values
labels, values = zip(*data)

# Create the bar plot
plt.figure(figsize=(20, 15))  # Set the figure size
plt.barh(labels, values, color='skyblue')  # Use horizontal bar plot

# Add titles and labels
plt.xlabel('Frequency')
plt.title('Frequency of Data-Related Terms')

# Show the plot
plt.tight_layout()  # Adjust layout for better fit
plt.show()
```
